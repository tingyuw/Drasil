# Introduction
A pendulum consists of mass attached to the end of a rod, its moving curve is highly sensitive to initial conditions Therefore, it is useful to have a program the to simulate the motion of the pendulum to exhibit the chaotic characteristics of it The program documented here is called pendulum.
The following section provides an overview of the Software Requirements Specification (SRS) for Pendulum. This section explains the purpose of this document, the scope of the requirements, the characteristics of the intended reader, and the organization of the document.
<div id="Sec:ReqsScope">

## Scope of Requirements
The scope of the requirements includes the analysis of a two-dimensional (2D) pendulum motion problem with various initial conditions..
</div>
</div>

#### Detailed derivation of angular frequency pendulum:
Consider the torque, on a pendulum defined in <a href=#TM:NewtonSecLawRotMot>TM: NewtonSecLawRotMot</a>, the force providing the restoring torque is the component of the the weight of the pendulum bob that acts along the arc length The the torque is the length of the string the <em>L<sub>rod</sub></em> multiplied by the component of the net the force that is perpendicular to the radius of the arc. The minus sign indicates the torque acts in the opposite direction of the angular displacement:
<div id="">
$\mathbf{τ}=-{L_{\text{rod}}} m \mathbf{g} \sin\left({θ_{p}}\right)$
</div>
So then
<div id="">
$\mathbf{I} α=-{L_{\text{rod}}} m \mathbf{g} \sin\left({θ_{p}}\right)$
</div>
Therefore,
<div id="">
$\mathbf{I} \frac{\,d\frac{\,d{θ_{p}}}{\,dt}}{\,dt}=-{L_{\text{rod}}} m \mathbf{g} \sin\left({θ_{p}}\right)$
</div>
Substituting for <em><b>I</b></em>
<div id="">
$m {L_{\text{rod}}}^{2} \frac{\,d\frac{\,d{θ_{p}}}{\,dt}}{\,dt}=-{L_{\text{rod}}} m \mathbf{g} \sin\left({θ_{p}}\right)$
</div>
Crossing out <em>m</em> and <em>L<sub>rod</sub></em> we have
<div id="">
$\frac{\,d\frac{\,d{θ_{p}}}{\,dt}}{\,dt}=-\left(\frac{\mathbf{g}}{{L_{\text{rod}}}}\right) \sin\left({θ_{p}}\right)$
</div>
For small angles, we approximate sin <em>θ<sub>p</sub></em> to <em>θ<sub>p</sub></em>
<div id="">
$\frac{\,d\frac{\,d{θ_{p}}}{\,dt}}{\,dt}=-\left(\frac{\mathbf{g}}{{L_{\text{rod}}}}\right) {θ_{p}}$
</div>
Because this equation, has the same form as the equation for simple harmonic motion the solution is easy to find.  The angular frequency
<div id="">
$Ω=\sqrt{\frac{\mathbf{g}}{{L_{\text{rod}}}}}$
</div>
</div>

#### Detailed derivation of angular displacement:
When the pendulum is displaced to an initial angle and released, the pendulum swings back and forth with periodic motion By applying Newton's Second Law for Rotation in <a href=#TM:NewtonSecLawRotMot>TM: NewtonSecLawRotMot</a>, the equation of motion for the pendulum may be obtained:
<div id="">
$\mathbf{τ}=\mathbf{I} α$
</div>
Where <em><b>τ</b></em> denotes the torque, <em><b>I</b></em> denotes the moment of inertia and <em>α</em> denotes the angular acceleration This implies:
<div id="">
$-m \mathbf{g} \sin\left({θ_{p}}\right) {L_{\text{rod}}}=m {L_{\text{rod}}}^{2} \frac{\,d\frac{\,d{θ_{p}}}{\,dt}}{\,dt}$
</div>
And rearranged as:
<div id="">
$\frac{\,d\frac{\,d{θ_{p}}}{\,dt}}{\,dt}+\frac{\mathbf{g}}{{L_{\text{rod}}}} \sin\left({θ_{p}}\right)=0$
</div>
If the amplitude of angular displacement is small enough, we can approximate <em>sin(θ<sub>p</sub>) = θ<sub>p</sub></em> for the purpose of a simple pendulum at very small angles. Then the equation of motion reduces to the equation of simple harmonic motion:
<div id="">
$\frac{\,d\frac{\,d{θ_{p}}}{\,dt}}{\,dt}+\frac{\mathbf{g}}{{L_{\text{rod}}}} {θ_{p}}=0$
</div>
Thus the simple harmonic motion is:
<div id="">
${θ_{p}}\left(t\right)={θ_{i}} \cos\left(Ω t\right)$
</div>
</div>
</div>
